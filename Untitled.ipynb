```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset
file_path = '/Users/benothmane/Downloads/student_health_data.csv'
data = pd.read_csv(file_path)

# Vérifier les valeurs uniques et leurs fréquences dans Health_Risk_Level
risk_level_counts = data['Health_Risk_Level'].value_counts()

# Diagramme en camembert pour Health_Risk_Level
plt.figure(figsize=(8, 8))
plt.pie(
    risk_level_counts, 
    labels=risk_level_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("pastel")
)
plt.title('Répartition des niveaux du risque de santé des étudiants')
plt.axis('equal')  # Assure que le camembert est un cercle
plt.show()
```

```python
# Encodage des variables catégoriques (adapté pour d'autres colonnes)
df_encoded = df.copy()

# Encoder les variables catégoriques existantes
df_encoded['Physical_Activity'] = df_encoded['Physical_Activity'].map({'Low': 0, 'Moderate': 1, 'High': 2})
df_encoded['Sleep_Quality'] = df_encoded['Sleep_Quality'].map({'Poor': 0, 'Moderate': 1, 'Good': 2})
df_encoded['Mood'] = df_encoded['Mood'].map({'Stressed': 0, 'Happy': 1, 'Neutral': 2})
df_encoded['Health_Risk_Level'] = df_encoded['Health_Risk_Level'].map({'Low': 0, 'Moderate': 1, 'High': 2})

# Si des variables catégoriques supplémentaires sont présentes (comme Gender), les encoder
if 'Gender' in df_encoded.columns:
    df_encoded['Gender'] = df_encoded['Gender'].astype('category').cat.codes

# Variables à inclure dans la heatmap
variables_to_include = [
    'Physical_Activity', 'Sleep_Quality', 'Mood', 'Health_Risk_Level', 
    'Heart_Rate', 'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic',
    'Stress_Level_Biosensor', 'Stress_Level_Self_Report', 'Study_Hours', 'Project_Hours'
]

# Filtrer le DataFrame pour les variables sélectionnées
df_corr = df_encoded[variables_to_include]

# Calcul de la matrice de corrélation
corr_matrix = df_corr.corr()

# Visualiser la heatmap des corrélations
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrice de corrélation des indicateurs  sur Health_Risk_Level', fontsize=14)
plt.show()

# Focus sur Health_Risk_Level (corrélations spécifiques)
health_risk_corr = corr_matrix['Health_Risk_Level'].sort_values(ascending=False)
print("\nCorrélations avec Health_Risk_Level :\n", health_risk_corr)

# Visualiser uniquement les corrélations avec Health_Risk_Level
plt.figure(figsize=(6, 8))
sns.barplot(x=health_risk_corr.values, y=health_risk_corr.index, palette='viridis')
plt.title("Corrélations entre les indicateurs sur Health_Risk_Level")
plt.xlabel("Coefficient de corrélation")
plt.ylabel("Variables")
plt.show()
```
