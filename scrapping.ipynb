```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page Amazon
url = "https://www.amazon.fr/s?k=livres&__mk_fr_FR=ÅMÅŽÕÑ&crid=18QYHZLRIFKL5&sprefix=livres%2Caps%2C75&ref=nb_sb_noss_1"

# En-têtes HTTP pour simuler un navigateur
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36",
    "Accept-Language": "fr-FR,fr;q=0.9,en-US;q=0.8,en;q=0.7",
}

# Envoyer la requête HTTP
response = requests.get(url, headers=headers)

# Vérifier si la requête a réussi
if response.status_code == 200:
    print("Page fetched successfully!")
else:
    print("Failed to fetch the page. Status code:", response.status_code)
    exit()

# Analyser le contenu HTML
soup = BeautifulSoup(response.content, 'lxml')

# Trouver tous les conteneurs de livres
books = soup.find_all('div', {'data-component-type': 's-search-result'})

# Extraire les titres et les prix
data = []
for book in books:
    title_element = book.find('span', class_='a-size-base-plus a-color-base a-text-normal')
    price_element = book.find('span', class_='a-offscreen')
    
    if title_element and price_element:
        title = title_element.get_text(strip=True)
        price = price_element.get_text(strip=True).replace('\xa0', ' ')
        data.append({'Title': title, 'Price': price})

# Créer un DataFrame à partir des données extraites
df = pd.DataFrame(data)

# Afficher le DataFrame
df

```
